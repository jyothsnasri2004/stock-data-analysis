```python
import yfinance as yf
import pandas as pd

tickers = ["TCS.NS","RELIANCE.NS"]
data = yf.download(tickers, period = '5y' , auto_adjust = True)

for ticker in tickers:
  data[('daily returns',ticker)] = data[('Close',ticker)].pct_change()
  data[('ma20',ticker)] = data[('Close',ticker)].rolling(window = 20).mean()
  data[('ma50',ticker)] = data[('Close',ticker)].rolling(window = 50).mean()
  data[('volitality',ticker)] = data[('Close',ticker)].rolling(window = 20).std()

data.head()
```

```python
#a two-sample t-test( closing mean of tcs = reliance or not)

from scipy.stats import ttest_ind

t_test, p_value = ttest_ind(data[('Close','TCS.NS')],data[('Close','RELIANCE.NS')])
print(f"t-test: {t_test} , p-value:{p_value}")

if p_value < 0.05:
  print("reject null hypo i.e both are not significantly same")
else:
  print("accept i.e both are singnificantly same")
```
